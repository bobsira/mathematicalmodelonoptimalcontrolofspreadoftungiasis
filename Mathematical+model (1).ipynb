```python
def make_chart_simple_line_chart():

    years = [1,2,3,4]
    gdp = [239, 167, 103, 94]

    # create a line chart, years on x-axis, gdp on y-axis
    plt.plot(years, gdp, color='red', marker='o', linestyle='solid')

    # add a title
    #plt.title("Nominal GDP")

    # add a label to the y-axis
    plt.ylabel("Household Members Susceptible Suffer Jigger Infestation")
    plt.xlabel("The week in which the reading was recorded")
    plt.show()
```
